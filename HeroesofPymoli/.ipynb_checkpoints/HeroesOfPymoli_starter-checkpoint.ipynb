## Heroes of PyMoli

### Heroes of PyMoli - Analysis
* The majority of players are male, with 84%.
* Female players are more likely to buy expensive things and bring more value per purchase than male players.
* The most populous age group is 20-24 years old(44.8 %), followed by 15-19 years old(18.6%) and 25-29 years old(13.4 %).
* The average purchase is around USD 3 per individual.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# csv file to load
file = r"C:/Users/shrey/Documents/pandas-challenge/HeroesofPymoli/Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
purchase_data.head()
```

## Player Count

* Calculate the total number of players

```python
# Count number of unique SN
total_players = purchase_data["SN"].nunique()

# Save as a Data Frame for cleaner formatting
total_players = pd.DataFrame({"Total number of players:":[total_players]})

# Display the summary
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
# Number of unique items
unique_items = purchase_data["Item ID"].nunique()

# Average purchase price
purchase_price = purchase_data["Price"].mean()

# Number of purchases
total_purchase = purchase_data["Purchase ID"].count()

# Total Revenue
total_revenue = purchase_data["Price"].sum()
```

```python
# Create a summary data frame to hold the results
summary_df = pd.DataFrame({
                            "Number of Unique Items":[unique_items],
                            "Average Price":[purchase_price],
                            "Total Number of Purchases":[total_purchase],
                            "Total Revenue":[total_revenue]
                          })

# Give the displayed data cleaner formatting
summary_df["Average Price"] = summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)

# Display the summary data frame
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Count of Male, Female and Other/Non-Disclosed Players
# Data frame for players of gender
gender_df = purchase_data.groupby(["Gender"])

# Total count of Male, Female and other players
gender_count = gender_df["SN"].nunique()
# Percentage of Male, Female and other players out of total players
gender_percentage = (gender_count / purchase_data["SN"].nunique()) * 100

# Summary data frame - Gender
# Create a summary data frame to hold the results
gender_summary_df = pd.DataFrame({"Total Gender Count": gender_count,
                                   "Percentage of Players": gender_percentage})

# Give the displayed data cleaner formatting
gender_summary_df["Percentage of Players"] = gender_summary_df["Percentage of Players"].map("{:,.2f}%".format)

# Display the summary table and sort
gender_summary_df = gender_summary_df.sort_values(["Total Gender Count"], ascending=False)
gender_summary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
# Number of unique items by gender
purchase_count = purchase_data.groupby("Gender")["Item ID"].count()

# Average purchase price by gender
purchase_gender_price = purchase_data.groupby("Gender")["Price"].mean()

# Number of purchases by gender
total_gender_purchase = purchase_data.groupby("Gender")["Purchase ID"].sum()

# Average Purchase Total per Person by gender
average_purchase = total_gender_purchase / gender_count
```

```python
# Create a summary data frame to hold the results
summary_gender_df = pd.DataFrame({
                            "Number of Unique Items by Gender":purchase_count,
                            "Average Price by Gender":purchase_gender_price,
                            "Total Number of Purchases by Gender":total_gender_purchase,
                            "Average Purchase Total per Person by Gender":average_purchase
                          })

# Give the displayed data cleaner formatting
summary_gender_df["Average Price by Gender"] = summary_gender_df["Average Price by Gender"].map("${:,.2f}".format)
summary_gender_df["Average Purchase Total per Person by Gender"] = summary_gender_df["Average Purchase Total per Person by Gender"].map("${:,.2f}".format)

# Display the summary data frame
summary_gender_df
```

## Age Demographics

* Establish bins for ages
* Categorize the existing players using the age bins. Hint: use pd.cut()
* Calculate the numbers and percentages by age group
* Create a summary data frame to hold the results
* Optional: round the percentage column to two decimal points
* Display Age Demographics Table

```python
# Establish bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Create labels for these bins
age_group =  ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Slice the data and place it into bins
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=age_group)
grouped_age_df = purchase_data.groupby(["Age Range"])

# Calculate numbers and percentages by age group
age_number = grouped_age_df["SN"].nunique()
age_percentage = ( age_number / purchase_data["SN"].nunique()) * 100
```

```python
# Summary data frame - Age
# Create a summary data frame to hold the results
age_summary_df = pd.DataFrame({"Total Age Count": age_number,
                                "Percentage of Players - Age": age_percentage})

# Give the displayed data cleaner formatting
age_summary_df["Percentage of Players - Age"] = age_summary_df["Percentage of Players - Age"].map("{:,.2f}%".format)

# Display the summary table and sort
age_summary_df = age_summary_df.sort_values(["Age Range"], ascending=True)
age_summary_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age
* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc.
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
# Number of unique items by age
count_purchase = grouped_age_df["Purchase ID"].count()

# Average purchase price by age
avg_purchase_price = grouped_age_df["Price"].mean()

# Total Purchase value by age
total_purchase_value = grouped_age_df["Price"].sum()

# Average Purchase Total per Person by age
avg_total = total_purchase_value / age_number
```

```python
# Create a summary data frame to hold the results
age_df = pd.DataFrame({     "Number of Unique Items by Age":count_purchase,
                            "Average Price by Age":avg_purchase_price,
                            "Total Purchase Value by Age":total_purchase_value,
                            "Average Purchase Total per Person by Age":avg_total
                          })

# Give the displayed data cleaner formatting
age_df["Average Price by Age"] = age_df["Average Price by Age"].map("${:,.2f}".format)
age_df["Total Purchase Value by Age"] = age_df["Total Purchase Value by Age"].map("${:,.2f}".format)
age_df["Average Purchase Total per Person by Age"] = age_df["Average Purchase Total per Person by Age"].map("${:,.2f}".format)

# Display the summary data frame
age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Top 5 spenders in the game by total purchase value
# Data frame for top spenders
grouped_spenders_df = purchase_data.groupby(["SN"])

# Count of SN
#SN_count = grouped_spenders_df["SN"].nunique()

# Purchase count of top 5 spenders
purchaseCount = grouped_spenders_df["Purchase ID"].count()

# Average Purchase Price
averagePurchasePrice = grouped_spenders_df["Price"].mean()

# Total Purchase Valye
totalPurchaseValue = grouped_spenders_df["Price"].sum()

# Summary data frame - Top 5 spenders
# Create a summary data frame to hold the results
spenders_summary_df = pd.DataFrame({"Purchase Count": purchaseCount,
                                   "Average Purchase Price": averagePurchasePrice,
                                   "Total Purchase Value": totalPurchaseValue})

# Give the displayed data cleaner formatting
spenders_summary_df["Average Purchase Price"] = spenders_summary_df["Average Purchase Price"].map("${:,.2f}".format)
spenders_summary_df["Total Purchase Value"] = spenders_summary_df["Total Purchase Value"].map("${:,.2f}".format)

# Display the summary table and sort
spenders_summary_df = spenders_summary_df.sort_values(["Total Purchase Value"], ascending=False)
spenders_summary_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name, and Item Price columns
popular_items_df = purchase_data[["Item ID", "Item Name", "Price"]]

# Group by Item ID and Item Name. 
grouped_popular_items_df = popular_items_df.groupby(["Item ID", "Item Name"])

#Perform calculations to obtain purchase count, average item price, and total purchase value
# Purchase count of most popular items
itemscount = grouped_popular_items_df["Item ID"].count()

# Average Purchase Price
itemprice = grouped_popular_items_df["Price"].mean()

# Total Purchase Valye
itempurchasevalue = grouped_popular_items_df["Price"].sum()

# Summary data frame - Top 5 spenders
# Create a summary data frame to hold the results
popular_items_df = pd.DataFrame({"Purchase Count": itemscount,
                                   "Item Price": itemprice,
                                   "Total Purchase Value": itempurchasevalue})

# Give the displayed data cleaner formatting
final_popular_df = popular_items_df.sort_values(["Purchase Count"], ascending=False)
final_popular_df["Item Price"] = popular_items_df["Item Price"].astype(float).map("${:,.2f}".format)
final_popular_df["Total Purchase Value"] = popular_items_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)

# Display the summary table
final_popular_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Create a data frame for most profitable items and sort based on Total Purchase Value
profitable_items_df = popular_items_df.sort_values(["Total Purchase Value"], ascending=False)

# Give the displayed data cleaner formatting
profitable_items_df["Item Price"] = profitable_items_df["Item Price"].astype(float).map("${:,.2f}".format)
profitable_items_df["Total Purchase Value"] = profitable_items_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)

# Display the data fram
profitable_items_df.head()
```
